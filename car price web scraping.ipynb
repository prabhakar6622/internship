```python
#importing libraries
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException,NoSuchElementException,NoSuchWindowException  
import requests
import re
```

```python
# lets first connect to the web driver
driver=webdriver.Chrome(r"C:/Users/Jai Mata Di/Downloads/chromedriver_win32/chromedriver.exe")
```

```python
#opening the cars24 website
url = 'https://www.cars24.com/'
driver.get(url)
```

```python
driver.find_element_by_xpath("//span[@class='_1Aa4N']").click()
```

```python
#Creating empty lists
brand=[]
year=[]
fuel=[]
transmission=[]
km=[]
price=[]
prod_URL=[]
```

```python
#Getting the URL of the cars
for i in range(0,500):
    url=driver.find_elements_by_xpath("//div[@class='col-4']/a")
    for i in url:
        prod_URL.append(i.get_attribute('href'))
   
        pass
   
```

```python
prod_URL
```

```python
for i in prod_URL:
    driver.get(i) 
    time.sleep(2)  
#Extracting car name
    try:
        brand.append(driver.find_element_by_xpath("//h2[@class='_3FpCg']").text)
    except NoSuchElementException as e:
        brand.append("-")
    
        
#Extracting the transmission
    try:
        transmission.append(driver.find_element_by_xpath("//p[@class='cvakB']/span").text)  
    except NoSuchElementException as e:
        transmission.append("-")
        
#Extracting the fuel consumed
    try:
        fuel.append(driver.find_element_by_xpath("//ul[@class='bVR0c']/li[3]").text)
    except NoSuchElementException as e:
        fuel.append("-")
        
#Extracting km driven
    try:
        km.append(driver.find_element_by_xpath("//ul[@class='bVR0c']/li[1]").text)
    except NoSuchElementException as e:
        km.append("-")
        

        #Extracting the price details
    try:
        price.append(driver.find_element_by_xpath("//div[@class='_7udZZ']").text.replace('',''))
    except NoSuchElementException as e:
        price.append("-")
```

```python
#Creating a dataframe and checking the data extracted
cars24 = pd.DataFrame({'name':brand,'selling_price':price,'km_driven':km,'fuel':fuel,'transmission':transmission})
cars24
```

```python
#Saving in a csv file
cars24.to_csv('cars24data2022.csv')
pd.read_csv('cars24data2022.csv')
```
