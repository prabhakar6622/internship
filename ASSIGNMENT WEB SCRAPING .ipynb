```python
#Q2.Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.
```

```python
page=requests.get('https://www.imdb.com/chart/top/')
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
movies=[]
for i in soup.find_all('td',class_='titleColumn'):
    movies.append(i.text.replace('\n',""))
    
movies
```

```python
#find rating of movies on imdb movies websites
```

```python
rating=[]
for i in soup.find_all('td',class_='ratingColumn imdbRating'):
    rating.append(i.text.replace('\n',""))

rating
```

```python
#find year of movies in imdb websites
```

```python
year=[]
for i in soup.find_all('span','secondaryInfo'):
    year.append(i.text)
    
year
```

```python
#make dataframe 
```

```python
import pandas as pd 
df=pd.DataFrame({'movies':movies,'rating':rating,'year':year})
df
```
